```r
#1 Constant Function
f<- function(x){
    fx<- 1
    return (fx)
}
    input<-0:5
    plot (input,sapply(input,f),type="l" ,xlab="x"
         ,ylab="f(x)")
```

```r
#2 Linear function

f<- function(x){
    fx<-5*x+10
    return (fx)
}
    input<-0:20
    plot (input,sapply(input,f),type="l" ,xlab="x"
         ,ylab="f(x)")
```

```r
#3 quadratic function
f<- function(x){
    fx<- 2*x^2+3*x+2
    return (fx)
}
    input<--20:20
    plot (input,sapply(input,f),type="l" ,xlab="x"
         ,ylab="f(x)")
```

```r
#4 polynomial function
f<- function(x){
    fx<- x^3-5*x^2+x-3
    return (fx)
}
    input<-seq (-12,10,0.2)
    plot (input,sapply(input,f),type="l" ,xlab="x"
         ,ylab="f(x)")
```

```r
#5 rational function
f<- function(x){
    fx<- 1/x
    return (fx)
}
    input<-seq (2,13,0.4)
    plot (input,sapply(input,f),type="l" ,xlab="x"
         ,ylab="f(x)")
```
